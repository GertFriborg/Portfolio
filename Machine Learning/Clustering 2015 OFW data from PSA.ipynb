# Dataset Description


## Description of Dataset

* The dataset that was chosen is the Survey on Overseas Filipino Workers

  The purpose of the survey was to collect data about the number of overseas workers in the Philippines, their socioeconomic status, and other relevant details about foreign workers who were employed between April and September of 2015.

## Collection of Data
The data collection for the 2015 Survey on Overseas Filipinos (SOF) involved face-to-face interviews, leveraging the 2003 Master Sample, which is designed for household surveys. The survey targeted Filipino workers who had worked abroad between April and September 2015, focusing on their socio-economic characteristics and remittances. Using the Labor Force Survey as a sampling base, it provided national and regional estimates. Surpervised by a Supervision Team consisting of Regional Directors (RDs), Provincial Statistical Officers (PSOs) and a three-stage sampling process. The survey used a stratified, multistage sampling design. It focuses on 17 administrative regions in the Philippines, using area samples (Primary Sampling Units) selected with probability proportional to size. Both were done to ensure data quality, but the sample's limitation to private households excludes institutional populations.

The two questionnares that were used are:

SOF Form 1: Two-page questionnaire to gather data on the socio-economic characteristics of overseas Filipinos (OFs) and their remittances. Each overseas member received a separate form if multiple members were abroad during the reference period.

SOF Form 2: Household control form, used to determine if households have members who are overseas.

For the numerical variables, we will be imputing the missing variables with the median. However, for RQ22_CASHAMT specifically (due to high number of missing values) we must first compare the distribution of the original data against the data with missing values imputed with the median. If the distributions are similar, we can proceed with imputing the missing values for the median. Otherwise, we will instead drop the missing values.

```python
df_imputed = df.copy()
df_imputed['RQ22_CASHAMT'].fillna(df_imputed['RQ22_CASHAMT'].median(), inplace=True)
# Original Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['RQ22_CASHAMT'], kde=True, color='blue', label='Original Distribution')
plt.title('Original Distribution of RQ22_CASHAMT')
plt.xlabel('RQ22_CASHAMT')
plt.ylabel('Frequency')

# Imputed Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df_imputed['RQ22_CASHAMT'], kde=True, color='red', label='Imputed Distribution')
plt.title('Distribution of RQ22_CASHAMT Imputed with Median')
plt.xlabel('RQ22_CASHAMT')
plt.ylabel('Frequency')

plt.show()


original_mean = df['RQ22_CASHAMT'].mean()
original_median = df['RQ22_CASHAMT'].median()
original_mode = df['RQ22_CASHAMT'].mode()[0]  # Mode can return multiple values, take the first one

print("Original Distribution:")
print(f"Mean: {original_mean}")
print(f"Median: {original_median}")
print(f"Mode: {original_mode}")


imputed_mean = df_imputed['RQ22_CASHAMT'].mean()
imputed_median = df_imputed['RQ22_CASHAMT'].median()
imputed_mode = df_imputed['RQ22_CASHAMT'].mode()[0]  # Mode can return multiple values, take the first one

print("\nImputed Distribution:")
print(f"Mean: {imputed_mean}")
print(f"Median: {imputed_median}")
print(f"Mode: {imputed_mode}")
```

We can see that there are differences in the mean and mode between the original data and the imputed data. As such, we will proceed with dropping missing values instead of imputing the median for RQ22_CASHAMT, sacrificing data size but keeping the natural frequency pattern of the original dataset.

```python
#addressing the null values for RQ16_NOMONTH, RQ22_CASHAMT we will replace those values with the median, being safe with the possibility of outliers and skewed distributiosn

df['RQ16_NOMONTH'].fillna(df['RQ16_NOMONTH'].median(), inplace=True)
df=df.dropna(subset=['RQ22_CASHAMT'])
```

For RQ8_HGRADE

```python
def convert_education_level(grade):
    # Extract the first digit as an integer
    first_digit = int(str(grade)[0])

    # Classify based on the first digit
    if first_digit == 0:
        return 'No Grade Completed'
    elif first_digit == 1:
        return 'Elementary Undergraduate'
    elif first_digit == 2:
        return 'Elementary Graduate'
    elif first_digit == 3:
        return 'High School Undergraduate'
    elif first_digit == 4:
        return 'High School Graduate'
    elif first_digit == 5:
        return 'Post Secondary Non-Tertiary Technical-Vocational Undergraduate'
    elif first_digit == 6:
        return 'College Undergraduate'
    else:
        return 'Unknown'  # Handle unexpected values

df['RQ8_HGRADE'] = df['RQ8_HGRADE'].apply(convert_education_level)
```

We are going to use K-means as a way to cluster and let's take a look at each cluster information.

First we must determine what K-value to use. We can do this by utilizing the elbow method.

```python
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(df_encoded)
  wcss.append(kmeans.inertia_)  # Inertia is the WCSS

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

We can see that the curve starts to inflect at K=4 so we will set our K to that.

WCSS = Within Cluster sum of squares

```python
kmeans = KMeans(n_clusters=4, random_state=42)  # Adjust n_clusters as needed
df_encoded['Cluster'] = kmeans.fit_predict(df_encoded)

cluster_summary = df_encoded.groupby('Cluster').mean()
cluster_summary
```

Now we look at how many observations or data points are in each cluster. With Cluster 2 being the highest amount.

```python
print("\nCluster Distribution:")
print(df_encoded['Cluster'].value_counts())
```

```python
# Apply PCA to reduce the data to 2D for visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_encoded.drop('Cluster', axis=1))

plt.figure(figsize=(18, 16))
plt.scatter(pca_components[:, 0], pca_components[:, 1], c=df_encoded['Cluster'], cmap="magma")
plt.title("PCA of Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')

#plt.xlim([-5, 10])  # Example limits, adjust as needed
#plt.ylim([-5, 10])  # Example limits, adjust as needed
plt.show()
```

```python
loadings = pca.components_
loadings_df = pd.DataFrame(loadings.T, columns=['PC1', 'PC2'], index=df_encoded.drop('Cluster', axis=1).columns)
```

# Statistical Inference

The following libraries will be used to help with the Statistical Inference section.

```python
from scipy.stats import f_oneway, f
from sklearn.metrics import mean_squared_error
```

To know if the difference between clusters is significant we can test for the mean cash amount of each cluster and compare them accordingly



• 𝐻0: there is no difference in the mean values of cash amount in clusters 0, 1, 2, and 3

• 𝐻𝐴: At least one of the means is significantly different from the other


First we need to calculate the MSG(Mean Square between Groups) and MSE (Mean Standard Error) Before proceeding

First, MSG(Mean Square between Groups)

```python
overall_mean = df_encoded['RQ22_CASHAMT_standardized'].mean()

cluster_means = df_encoded.groupby('Cluster')['RQ22_CASHAMT_standardized'].mean()

cluster_sizes = df_encoded['Cluster'].value_counts()

msg = sum(cluster_sizes.loc[cluster] * (cluster_means.loc[cluster] - overall_mean) ** 2 for cluster in cluster_means.index) / (len(cluster_means) - 1)

print("\nMean Square Between (MSB):")
print(msg)
```

Next, we compute the MSE (Mean Standard Error)

```python
cluster_means = df_encoded.groupby('Cluster')['RQ22_CASHAMT_standardized'].mean()

mse_values = {}
for cluster in range(4):
    cluster_data = df[df_encoded['Cluster'] == cluster]['RQ22_CASHAMT_standardized']
    mse = mean_squared_error(cluster_data, [cluster_means[cluster]] * len(cluster_data))
    mse_values[cluster] = mse

print("\nMean Square Error (MSE) for each cluster:")
for cluster, mse in mse_values.items():
    print(f"Cluster {cluster}: {mse}")
```

Now that we have these values we can compute for the f-statistic



```python


cluster_0 = df_encoded[df_encoded['Cluster'] == 0]['RQ22_CASHAMT_standardized']
cluster_1 = df_encoded[df_encoded['Cluster'] == 1]['RQ22_CASHAMT_standardized']
cluster_2 = df_encoded[df_encoded['Cluster'] == 2]['RQ22_CASHAMT_standardized']
cluster_3 = df_encoded[df_encoded['Cluster'] == 3]['RQ22_CASHAMT_standardized']

anova_result = f_oneway(cluster_0, cluster_1, cluster_2, cluster_3)
print("\nANOVA Result for RQ22_CASHAMT_standardized:")
print(f"F-statistic: {anova_result.statistic}, p-value: {anova_result.pvalue}")


alpha = 0.05
if anova_result.pvalue < alpha:
    print("\nThe differences between clusters are statistically significant.")
else:
    print("\nThe differences between clusters are not statistically significant.")


means = [cluster_0.mean(), cluster_1.mean(), cluster_2.mean(), cluster_3.mean()]
clusters = ['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3']

plt.figure(figsize=(10, 6))
plt.bar(clusters, means, color=['blue', 'orange', 'green', 'red'])
plt.title('Mean Values of Clusters')
plt.xlabel('Clusters')
plt.ylabel('Mean Value of CashAMT')
plt.show()


df_degrees_of_freedom = len(df_encoded) - 1
x = np.linspace(0, 5, 1000)
y = f.pdf(x, dfn=3, dfd=df_degrees_of_freedom)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='F-distribution', color='blue')
plt.title('F-Distribution')
plt.xlabel('F-Value')
plt.ylabel('Density')
plt.legend()
plt.show()

```

With these findings we can safely reject the null hypothesis at a 95% confidence interval and conclude that there is at least one cluster with significant diffence in mean cash amount remittance among the clusters 0, 1, 2, 3.
